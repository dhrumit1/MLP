```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
print(sns.get_dataset_names())
```

```python
iris=sns.load_dataset("iris")
iris
```

```python
iris.columns
```

```python
plt.plot(iris["sepal_length"],iris["sepal_width"])
```

```python
iris.plot(kind="scatter",x="sepal_length",y="sepal_width")
plt.grid()
plt.show()
```

```python
sns.set(style="darkgrid")
sns.scatterplot(x="sepal_length",y="sepal_width",data=iris)
```

```python
sns.set(style="darkgrid")
sns.scatterplot(x="sepal_length",y="petal_width",data=iris,hue="species")
```

```python
cen=pd.read_csv("census_data.csv")
cen
```

```python
cen.columns
```

```python
cen.plot(kind="scatter",x="age",y="education_num")
plt.grid()
plt.show()
```

```python
cen.plot(kind="scatter",x="age",y="capital_gain")
plt.grid()
plt.show()
```

```python
sns.relplot(x="capital_loss",y="capital_gain",hue="marital_status",data=cen)
```

```python
sns.catplot(x="age",y="marital_status",kind="box",hue="gender",data=cen)
```

```python
sns.catplot(x="age",y="marital_status",kind="bar",hue="gender",data=cen)
```

```python
sns.catplot(x="age",y="marital_status",kind="violin",hue="gender",data=cen)
```

```python
sns.catplot(x="marital_status",kind="count",hue="gender",data=cen)
```
