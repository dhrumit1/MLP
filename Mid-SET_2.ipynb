```python
#Q6  Create a line chart which shows proportion for each car brand.
car = cars['brand'].unique()
num_car = []
for brand in car:
    num = len(cars[cars['brand']==brand])
    num_car.append(num)
print(num_car)
fig = plt.figure(figsize =(10, 7))
plt.pie(num_car, labels = car)
 
# show plot
plt.show()
```
