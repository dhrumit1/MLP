```python
#Q2 Draw the bar chart of costliest and cheapest diamonds . 
data=diamond["price"]
data
maxprice=max(data)
minprice=min(data)
print(maxprice,minprice)
```

```python
sort=diamond.sort_values("price")
sort

# Select the costliest and cheapest diamonds
top=sort.head(1)
bottom=sort.tail(1)

# Concatenate the top and bottom diamonds into a single DataFrame
merge=pd.concat([top,bottom])

# Create a bar chart of the selected diamonds
merge.plot(x="carat", y="price", kind="bar", figsize=(8,6), grid=True)
```

```python
#Q3 Draw the chart for Price vs Carat of Diamonds. 
x=diamond["price"]
y=diamond["carat"]
plt.plot(x,y,"b:")
plt.show()
```
