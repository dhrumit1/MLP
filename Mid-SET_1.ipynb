```python
revlist=[]
revdic={}
for year in uyear:
    ryear=movieData[movieData["Year"]==year]
    revofyear=ryear["Revenue (Millions)"]
    print(f"Total Revenue of the year{year} is {revofyear.sum()} in Millions")
    revlist.append(revofyear.sum())
    revdic[year]=revofyear.sum()
```

```python
#Q6 In which year we had the most movies making profits?
revlist
```

```python
maxrev=revlist[0]
for revnue in revlist:
    if maxrev<revnue:
        maxrev=revnue
print(maxrev)
```
