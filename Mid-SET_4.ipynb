```python
#Q5 Create a data frame containing total number of flights going to a particular destination.

destinations = flight['dest'].unique()
destinations
```

```python
destdict={}
for destination in destinations:
    counts = len(flight[flight['dest']==destination])
    destdict[destination]=counts
    print(f"Destination is {destination} : {counts} no of flights")
```
